```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV
data = pd.read_csv("other/sql_outputs/02_01_query_1a.csv")
```

```python
# Create a bar plot with Seaborn
plt.figure(figsize=(10, 3))  # Optional: define figure size
sns.barplot(data=data, x="group", y="verification_success_rate", hue="method")
plt.title("Verification Success Rate by Group and Method")
plt.xlabel("Group")
plt.ylabel("Verification Success Rate")
plt.legend(title="Method")
plt.show()
```

```python
# line plot
plt.plot(data["group"], data["verification_success_rate"], marker='o')

# labels and title
plt.xlabel("Group")
plt.ylabel("Verification Success Rate")
plt.title("Verification Success Rate by Group")

# Display 
plt.show()
```

```python
# Group by 'method' and sum the 'verification_success_rate'
method_grouped = data.groupby("method")["verification_success_rate"].sum()

# Plotting pie chart
method_grouped.plot(kind="pie", autopct='%1.1f%%', startangle=90, legend=False)

# title
plt.title("Verification Success Rate by Method")

# Display 
plt.show()
```

```python
# Group by both 'group' and 'method' and sum the 'verification_success_rate'
group_method_grouped = data.groupby(["group", "method"])["verification_success_rate"].sum()

# Plotting pie chart
group_method_grouped.plot(kind="pie", autopct='%1.1f%%', startangle=90, legend=False)

# title
plt.title("Verification Success Rate by Group and Method")

# Display
```
